```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

def run():
    SPREAD = 100
    user_input = input("Input size in inches (default is 13): ")
    FIGSIZE = int(user_input) if user_input.strip() else 13
    plt.figure(figsize=(FIGSIZE, FIGSIZE))

    lines, stations = load_stations()

    pos = {station.station_name: (station.x * SPREAD, station.y * -SPREAD) for station in stations}

    graph = nx.Graph()
    for station in stations:
        graph.add_node(station.station_name, cn=station.station_line[0].colour)
        for connection in station.connections:
            graph.add_edge(station.station_name, connection.next_station.station_name, 
                        ec=connection.line.colour, st=connection.connection_style)
            
            edge = [(station.station_name, connection.next_station.station_name)]
            
            # White border
            nx.draw_networkx_edges(graph, pos, edgelist=edge, width=7, edge_color="white", style=connection.connection_style)
            # Fill color
            nx.draw_networkx_edges(graph, pos, edgelist=edge, width=5, edge_color=connection.line.colour, style=connection.connection_style)

    node_colors = list(nx.get_node_attributes(graph, 'cn').values())
    nx.draw_networkx_nodes(graph, pos, node_color=node_colors, edgecolors="white", linewidths=2)
    nx.draw_networkx_labels(graph, pos)

    plt.axis('equal')
    plt.show()

def load_stations():
    global data
    lines: dict[int, Line] = {}
    stations: dict[int, Station] = {}
    parsed = eval(data)
    # Parse lines into dict
    for line in parsed["lines"].values():
        lines[line["id"]] = (Line(**line))
    
    # Parse stations into dict without connections
    for station in parsed["stations"].values():
        stations[station["id"]] = Station(**{k: station[k] for k in ["id", "station_name", "x", "y", "lon", "lat"]})
        stations[station["id"]].station_line = link_lines_connected(lines, station["station_line"])

    #Append the connections    
    for station in stations.values():
        station.connections = create_connections(lines, stations, parsed["stations"][str(station.id)]["connections"])

    return list(lines.values()), list(stations.values())

def create_connections(lines: dict[int, Line], stations: dict[int, Station], connections: dict):
    connections_list: list[Connection] = []
    for connection in connections:
        toAdd = Connection(stations[connection["next_station"]], lines[connection["line_id"]], connection["connection_style"])
        connections_list.append(toAdd)
        
    return connections_list

def link_lines_connected(lines: dict[int, Line], station_lines: list[int]):
    line_list: list[Line] = []
    for station_line in station_lines:
        line_list.append(lines[station_line])
    return line_list

class Line:
    def __init__(self, id: int, code: str, name: str, colour: str):
        self.id = id
        self.code = code
        self.name = name
        self.colour = colour

class Connection:
    def __init__(self, next_station: Station, line: Line, connection_style: str):
        self.next_station = next_station
        self.line = line
        self.connection_style = connection_style

class Station:
    def __init__(self, id: int, station_name: str, x: float, y: float, lon: float, lat: float):
        self.id = id
        self.station_name = station_name
        self.x = x
        self.y = y
        self.lon = lon
        self.lat = lat
    connections: list[Connection] 
    station_line: list[Line]
    def __str__(self):
        return f"id:{self.id}, station_name:{self.station_name}," + \
        f"connections: [{",".join(c.next_station.station_name for c in self.connections)}] " + \
        f"and station_lines: [{",".join(l.code for l in self.station_line)}]"

data = '{"lines":{"1":{"id":1,"code":"TE","name":"Thomson-East Coast Line","colour":"#8c5d27"},"2":{"id":2,"code":"NS","name":"North-South Line","colour":"#c13e31"},"3":{"id":3,"code":"EW","name":"East-West Line","colour":"#4f924d"},"4":{"id":4,"code":"DT","name":"Downtown Line","colour":"#3653b1"},"5":{"id":5,"code":"CC","name":"Circle Line","colour":"#e2a33a"},"6":{"id":6,"code":"NE","name":"North East Line","colour":"#861ca4"},"7":{"id":7,"code":"CG","name":"Changi Airport Line","colour":"#4f924d"},"8":{"id":8,"code":"STE","name":"Sentosa Express","colour":"#2c2924"}},"stations":{"1":{"id":1,"station_name":"Bukit Panjang","x":0.249463000939964,"y":0.312415333361566,"lon":103.7595978435728,"lat":1.380745117305777,"connections":[{"next_station":27,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"2":{"id":2,"station_name":"Bras Basah","x":0.61586143909352,"y":0.722700823416926,"lon":103.8504742755075,"lat":1.296046213077239,"connections":[{"next_station":3,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"3":{"id":3,"station_name":"Esplanade","x":0.671118744151481,"y":0.782952538739741,"lon":103.8555678410694,"lat":1.293128813660232,"connections":[{"next_station":20,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"4":{"id":4,"station_name":"Nicoll Highway","x":0.770581893255811,"y":0.76191225719844,"lon":103.8635059552432,"lat":1.300169689708844,"connections":[{"next_station":5,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"5":{"id":5,"station_name":"Stadium","x":0.785883916194939,"y":0.719831694115839,"lon":103.8762262339748,"lat":1.303370226020246,"connections":[{"next_station":6,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"6":{"id":6,"station_name":"Mountbatten","x":0.79438504005001,"y":0.675838378165847,"lon":103.8832482927035,"lat":1.306045925660755,"connections":[{"next_station":7,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"7":{"id":7,"station_name":"Dakota","x":0.796085264821024,"y":0.634714191516942,"lon":103.8893072153439,"lat":1.308814187524259,"connections":[{"next_station":23,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"8":{"id":8,"station_name":"Tai Seng","x":0.746778746461613,"y":0.471173821355015,"lon":103.8882140677722,"lat":1.334620877230753,"connections":[{"next_station":9,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"9":{"id":9,"station_name":"Bartley","x":0.713624363426836,"y":0.433875140440891,"lon":103.8789086939245,"lat":1.343501061529334,"connections":[{"next_station":87,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"10":{"id":10,"station_name":"Lorong Chuan","x":0.603959865696421,"y":0.366928790082208,"lon":103.8629182999564,"lat":1.351738537912768,"connections":[{"next_station":111,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"11":{"id":11,"station_name":"Marymount","x":0.468791996400793,"y":0.375536177985467,"lon":103.8400589180772,"lat":1.347866072173078,"connections":[{"next_station":145,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"12":{"id":12,"station_name":"Farrer Road","x":0.321722553708067,"y":0.53142553667783,"lon":103.8079652307079,"lat":1.31749372589551,"connections":[{"next_station":13,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"13":{"id":13,"station_name":"Holland Village","x":0.308120755539953,"y":0.579244358362604,"lon":103.7957537701134,"lat":1.310711201332724,"connections":[{"next_station":24,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"14":{"id":14,"station_name":"one-north","x":0.30302008122691,"y":0.677751131033238,"lon":103.7877848105538,"lat":1.300074702884209,"connections":[{"next_station":15,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"15":{"id":15,"station_name":"Kent Ridge","x":0.313221429852996,"y":0.72174444698323,"lon":103.784490451682,"lat":1.293984785167014,"connections":[{"next_station":16,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"16":{"id":16,"station_name":"Haw Par Villa","x":0.326823228021109,"y":0.762868633632136,"lon":103.7820063970724,"lat":1.283055273541765,"connections":[{"next_station":17,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"17":{"id":17,"station_name":"Pasir Panjang","x":0.342975363345744,"y":0.793472679510391,"lon":103.7920301869204,"lat":1.276003232785054,"connections":[{"next_station":18,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"18":{"id":18,"station_name":"Labrador Park","x":0.366778510139942,"y":0.828858607557124,"lon":103.8023555536629,"lat":1.272106958054335,"connections":[{"next_station":19,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"19":{"id":19,"station_name":"Telok Blangah","x":0.398232668403705,"y":0.859462653435379,"lon":103.8098837309202,"lat":1.270721226211404,"connections":[{"next_station":88,"line_id":5,"connection_style":"solid"}],"station_line":[5]},"20":{"id":20,"station_name":"Promenade","x":0.742528184534077,"y":0.804949196714737,"lon":103.8601368844023,"lat":1.293862087876265,"connections":[{"next_station":4,"line_id":5,"connection_style":"solid"},{"next_station":4,"line_id":5,"connection_style":"solid"},{"next_station":25,"line_id":4,"connection_style":"solid"}],"station_line":[4,5]},"21":{"id":21,"station_name":"MacPherson","x":0.772282118026825,"y":0.518992643039789,"lon":103.8886152116773,"lat":1.326192777550482,"connections":[{"next_station":8,"line_id":5,"connection_style":"solid"},{"next_station":42,"line_id":4,"connection_style":"solid"}],"station_line":[4,5]},"22":{"id":22,"station_name":"Botanic Gardens","x":0.349776262429801,"y":0.475955703523492,"lon":103.8159141457915,"lat":1.32268461233639,"connections":[{"next_station":12,"line_id":5,"connection_style":"solid"},{"next_station":50,"line_id":4,"connection_style":"solid"}],"station_line":[4,5]},"23":{"id":23,"station_name":"Paya Lebar","x":0.791834702893488,"y":0.582113487663691,"lon":103.8927883241053,"lat":1.318215992150078,"connections":[{"next_station":21,"line_id":5,"connection_style":"solid"},{"next_station":56,"line_id":3,"connection_style":"solid"}],"station_line":[3,5]},"24":{"id":24,"station_name":"Buona Vista","x":0.302169968841403,"y":0.637583320818028,"lon":103.7904491753026,"lat":1.307309450770241,"connections":[{"next_station":14,"line_id":5,"connection_style":"solid"},{"next_station":64,"line_id":3,"connection_style":"solid"}],"station_line":[3,5]},"25":{"id":25,"station_name":"Bayfront","x":0.701722790029737,"y":0.857549900567988,"lon":103.8595968721875,"lat":1.28283490868894,"connections":[{"next_station":20,"line_id":5,"connection_style":"solid"},{"next_station":34,"line_id":4,"connection_style":"solid"}],"station_line":[4,5]},"26":{"id":26,"station_name":"Changi Airport","x":1.1055261731456,"y":0.51038525513653,"lon":103.9892766424138,"lat":1.356341421004803,"connections":[],"station_line":[7]},"27":{"id":27,"station_name":"Cashew","x":0.249463000939964,"y":0.334411991336561,"lon":103.764234480711,"lat":1.369833318596293,"connections":[{"next_station":28,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"28":{"id":28,"station_name":"Hillview","x":0.248612888554457,"y":0.356408649311558,"lon":103.7676882901031,"lat":1.362474237945068,"connections":[{"next_station":147,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"29":{"id":29,"station_name":"Beauty World","x":0.249463000939964,"y":0.402314718128941,"lon":103.7759842962261,"lat":1.341696149932801,"connections":[{"next_station":30,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"30":{"id":30,"station_name":"King Albert Park","x":0.250313113325471,"y":0.423354999670241,"lon":103.7845615847806,"lat":1.33597604551556,"connections":[{"next_station":31,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"31":{"id":31,"station_name":"Sixth Avenue","x":0.249463000939964,"y":0.444395281211542,"lon":103.7968481782287,"lat":1.331521301435717,"connections":[{"next_station":32,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"32":{"id":32,"station_name":"Tan Kah Kee","x":0.28856817067329,"y":0.476912079957188,"lon":103.8078008737065,"lat":1.326146158825493,"connections":[{"next_station":22,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"33":{"id":33,"station_name":"Rochor","x":0.608210427623956,"y":0.635670567950637,"lon":103.8531043356494,"lat":1.304214777033743,"connections":[{"next_station":77,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"34":{"id":34,"station_name":"Downtown","x":0.645615372586269,"y":0.870939170639725,"lon":103.8526465600109,"lat":1.279773455282965,"connections":[{"next_station":35,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"35":{"id":35,"station_name":"Telok Ayer","x":0.574205932203673,"y":0.823120348954951,"lon":103.8488416560614,"lat":1.282250664264074,"connections":[{"next_station":89,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"36":{"id":36,"station_name":"Fort Canning","x":0.496845705122527,"y":0.699747789008234,"lon":103.8444250315653,"lat":1.292830275945718,"connections":[{"next_station":37,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"37":{"id":37,"station_name":"Bencoolen","x":0.62776301249062,"y":0.687314895370193,"lon":103.8492717038725,"lat":1.297699117654433,"connections":[{"next_station":38,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"38":{"id":38,"station_name":"Jalan Besar","x":0.656666833597861,"y":0.652885343757156,"lon":103.8552961254143,"lat":1.304889894129293,"connections":[{"next_station":39,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"39":{"id":39,"station_name":"Bendemeer","x":0.694071778560173,"y":0.61176115710825,"lon":103.8630740059371,"lat":1.31385250562225,"connections":[{"next_station":40,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"40":{"id":40,"station_name":"Geylang Bahru","x":0.720425262510893,"y":0.582113487663691,"lon":103.8715556075693,"lat":1.321561698740251,"connections":[{"next_station":41,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"41":{"id":41,"station_name":"Mattar","x":0.74762885884712,"y":0.549596688918044,"lon":103.8837483873128,"lat":1.326463522167072,"connections":[{"next_station":21,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"42":{"id":42,"station_name":"Ubi","x":0.807136725832616,"y":0.484563091426752,"lon":103.8985855709163,"lat":1.329830770674731,"connections":[{"next_station":43,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"43":{"id":43,"station_name":"Kaki Bukit","x":0.836040546939857,"y":0.450133539813714,"lon":103.9091462762628,"lat":1.334921890342718,"connections":[{"next_station":44,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"44":{"id":44,"station_name":"Bedok North","x":0.862394030890577,"y":0.42144224680285,"lon":103.9178105260292,"lat":1.334422576551513,"connections":[{"next_station":45,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"45":{"id":45,"station_name":"Bedok Reservoir","x":0.912550661635496,"y":0.409965729598504,"lon":103.9334871898551,"lat":1.336554903328631,"connections":[{"next_station":46,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"46":{"id":46,"station_name":"Tampines West","x":0.961857179994907,"y":0.409965729598504,"lon":103.9389485037816,"lat":1.344999197514483,"connections":[{"next_station":76,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"47":{"id":47,"station_name":"Tampines East","x":1.06047021671373,"y":0.437700646175673,"lon":103.9426536818783,"lat":1.354193927606422,"connections":[{"next_station":48,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"48":{"id":48,"station_name":"Upper Changi","x":1.06047021671373,"y":0.466391939186538,"lon":103.9610430807089,"lat":1.340840474064327,"connections":[{"next_station":49,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"49":{"id":49,"station_name":"Expo","x":1.06047021671373,"y":0.511341631570225,"lon":103.9621610218642,"lat":1.335950971715691,"connections":[{"next_station":146,"line_id":4,"connection_style":"solid"},{"next_station":26,"line_id":7,"connection_style":"dashed"}],"station_line":[4,7]},"50":{"id":50,"station_name":"Stevens","x":0.425436264739932,"y":0.474999327089797,"lon":103.8263225771446,"lat":1.320937743414412,"connections":[{"next_station":129,"line_id":1,"connection_style":"solid"},{"next_station":112,"line_id":4,"connection_style":"solid"}],"station_line":[1,4]},"51":{"id":51,"station_name":"Pasir Ris","x":1.00521291165576,"y":0.361190531480035,"lon":103.9491369403491,"lat":1.372886236352008,"connections":[{"next_station":76,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"52":{"id":52,"station_name":"Simei","x":1.00351268688475,"y":0.438657022609369,"lon":103.953818676297,"lat":1.342661794533708,"connections":[{"next_station":75,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"53":{"id":53,"station_name":"Bedok","x":0.921051785490567,"y":0.462566433451756,"lon":103.9294042796889,"lat":1.32407464586546,"connections":[{"next_station":54,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"54":{"id":54,"station_name":"Kembangan","x":0.863244143276084,"y":0.496995985064793,"lon":103.9128992887463,"lat":1.321152108055446,"connections":[{"next_station":55,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"55":{"id":55,"station_name":"Eunos","x":0.828389535470293,"y":0.536207418846308,"lon":103.9032151869783,"lat":1.319529918801268,"connections":[{"next_station":23,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"56":{"id":56,"station_name":"Aljunied","x":0.757830207473205,"y":0.618455792144119,"lon":103.8824728877824,"lat":1.316487930304078,"connections":[{"next_station":57,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"57":{"id":57,"station_name":"Kallang","x":0.731476723522485,"y":0.647147085154983,"lon":103.8715870758614,"lat":1.311781770543804,"connections":[{"next_station":58,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"58":{"id":58,"station_name":"Lavender","x":0.704273127186258,"y":0.678707507466934,"lon":103.8630430629033,"lat":1.307169460472736,"connections":[{"next_station":77,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"59":{"id":59,"station_name":"Tanjong Pagar","x":0.501096267050063,"y":0.845117006929947,"lon":103.8469275802349,"lat":1.27737948810714,"connections":[{"next_station":91,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"60":{"id":60,"station_name":"Tiong Bahru","x":0.415234916113847,"y":0.75139211642779,"lon":103.8269905951461,"lat":1.285759455898818,"connections":[{"next_station":61,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"61":{"id":61,"station_name":"Redhill","x":0.388881432163127,"y":0.720788070549535,"lon":103.8169944585527,"lat":1.289370704673515,"connections":[{"next_station":62,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"62":{"id":62,"station_name":"Queenstown","x":0.362527948212407,"y":0.691140401104975,"lon":103.805584729896,"lat":1.295446256070099,"connections":[{"next_station":63,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"63":{"id":63,"station_name":"Commonwealth","x":0.337874689032701,"y":0.666274613828893,"lon":103.7987785726048,"lat":1.302077045803189,"connections":[{"next_station":24,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"64":{"id":64,"station_name":"Dover","x":0.257964124795035,"y":0.638539697251724,"lon":103.7783047152878,"lat":1.31141543344877,"connections":[{"next_station":65,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"65":{"id":65,"station_name":"Clementi","x":0.219709067447215,"y":0.639496073685419,"lon":103.7655701161836,"lat":1.314078414345001,"connections":[{"next_station":113,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"66":{"id":66,"station_name":"Chinese Garden","x":0.116845468800857,"y":0.638539697251724,"lon":103.733393135763,"lat":1.341565826332578,"connections":[{"next_station":67,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"67":{"id":67,"station_name":"Lakeside","x":0.0743398495255031,"y":0.638539697251724,"lon":103.720448529393,"lat":1.343983341262197,"connections":[{"next_station":68,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"68":{"id":68,"station_name":"Boon Lay","x":0.0343845674066698,"y":0.591677252000645,"lon":103.7061455479667,"lat":1.338420695657893,"connections":[{"next_station":69,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"69":{"id":69,"station_name":"Pioneer","x":0.0343845674066698,"y":0.553422194652826,"lon":103.6971553734113,"lat":1.33734462692044,"connections":[{"next_station":70,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"70":{"id":70,"station_name":"Joo Koon","x":0.0343845674066698,"y":0.517079890172398,"lon":103.6785207090113,"lat":1.327502084458654,"connections":[{"next_station":71,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"71":{"id":71,"station_name":"Gul Circle","x":0.0352346797921769,"y":0.475955703523492,"lon":103.6604441438593,"lat":1.31903226656851,"connections":[{"next_station":72,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"72":{"id":72,"station_name":"Tuas Crescent","x":0.0343845674066698,"y":0.438657022609369,"lon":103.6489587412671,"lat":1.320763817112224,"connections":[{"next_station":73,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"73":{"id":73,"station_name":"Tuas West Road","x":0.0343845674066698,"y":0.402314718128941,"lon":103.6392843447087,"lat":1.329872424139264,"connections":[{"next_station":74,"line_id":3,"connection_style":"solid"}],"station_line":[3]},"74":{"id":74,"station_name":"Tuas Link","x":0.0343845674066698,"y":0.363103284347426,"lon":103.6372457200215,"lat":1.340426899205142,"connections":[],"station_line":[3]},"75":{"id":75,"station_name":"Tanah Merah","x":0.989060776331134,"y":0.46161005701806,"lon":103.9457230415822,"lat":1.327061275814356,"connections":[{"next_station":53,"line_id":3,"connection_style":"solid"},{"next_station":49,"line_id":7,"connection_style":"dashed"}],"station_line":[3,7]},"76":{"id":76,"station_name":"Tampines","x":1.00521291165576,"y":0.393707330225681,"lon":103.9426536818783,"lat":1.354193927606422,"connections":[{"next_station":52,"line_id":3,"connection_style":"solid"},{"next_station":47,"line_id":4,"connection_style":"solid"}],"station_line":[3,4]},"77":{"id":77,"station_name":"Bugis","x":0.673669081308003,"y":0.71122430621258,"lon":103.8568906548671,"lat":1.29907884888673,"connections":[{"next_station":114,"line_id":3,"connection_style":"solid"},{"next_station":20,"line_id":4,"connection_style":"solid"}],"station_line":[3,4]},"78":{"id":78,"station_name":"Clarke Quay","x":0.518098514760205,"y":0.722700823416926,"lon":103.846665833855,"lat":1.289383851872467,"connections":[{"next_station":116,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"79":{"id":79,"station_name":"Farrer Park","x":0.570805482661644,"y":0.554378571086522,"lon":103.8547637677557,"lat":1.313543965975699,"connections":[{"next_station":80,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"80":{"id":80,"station_name":"Boon Keng","x":0.569955370276137,"y":0.513254384437616,"lon":103.8607068332875,"lat":1.317914816472307,"connections":[{"next_station":81,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"81":{"id":81,"station_name":"Potong Pasir","x":0.593758517070336,"y":0.47213019778871,"lon":103.8689664447029,"lat":1.33280135224287,"connections":[{"next_station":82,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"82":{"id":82,"station_name":"Woodleigh","x":0.626062787719605,"y":0.432918764007196,"lon":103.8710538041579,"lat":1.339867219941902,"connections":[{"next_station":87,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"83":{"id":83,"station_name":"Kovan","x":0.692371553789159,"y":0.357365025745253,"lon":103.8847217321222,"lat":1.360205623063266,"connections":[{"next_station":84,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"84":{"id":84,"station_name":"Hougang","x":0.717874925354371,"y":0.329630109168084,"lon":103.8929724009716,"lat":1.371946116552726,"connections":[{"next_station":85,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"85":{"id":85,"station_name":"Buangkok","x":0.740827959763063,"y":0.302851569024611,"lon":103.8928838367306,"lat":1.383409790989895,"connections":[{"next_station":86,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"86":{"id":86,"station_name":"Sengkang","x":0.765481218942769,"y":0.276073028881137,"lon":103.8957054566102,"lat":1.392370843631531,"connections":[{"next_station":92,"line_id":6,"connection_style":"solid"}],"station_line":[6]},"87":{"id":87,"station_name":"Serangoon","x":0.661767507910903,"y":0.39179457735829,"lon":103.873546495734,"lat":1.349727699514359,"connections":[{"next_station":10,"line_id":5,"connection_style":"solid"},{"next_station":83,"line_id":6,"connection_style":"solid"}],"station_line":[5,6]},"88":{"id":88,"station_name":"HarbourFront","x":0.441588400064566,"y":0.889110322879939,"lon":103.8225316516256,"lat":1.265259883734691,"connections":[{"next_station":119,"line_id":5,"connection_style":"dotted"},{"next_station":91,"line_id":6,"connection_style":"solid"},{"next_station":120,"line_id":8,"connection_style":"dashdot"}],"station_line":[5,6,8]},"89":{"id":89,"station_name":"Chinatown","x":0.499396042279049,"y":0.744697481391922,"lon":103.8444585417297,"lat":1.284736083637029,"connections":[{"next_station":36,"line_id":4,"connection_style":"solid"},{"next_station":78,"line_id":6,"connection_style":"solid"}],"station_line":[4,6]},"90":{"id":90,"station_name":"Little India","x":0.571655595047152,"y":0.596459134169123,"lon":103.8483043991358,"lat":1.307683536558851,"connections":[{"next_station":33,"line_id":4,"connection_style":"solid"},{"next_station":79,"line_id":6,"connection_style":"solid"}],"station_line":[4,6]},"91":{"id":91,"station_name":"Outram Park","x":0.444138737221088,"y":0.781996162306045,"lon":103.8403867870957,"lat":1.280213353869118,"connections":[{"next_station":133,"line_id":1,"connection_style":"solid"},{"next_station":60,"line_id":3,"connection_style":"solid"},{"next_station":89,"line_id":6,"connection_style":"solid"}],"station_line":[1,3,6]},"92":{"id":92,"station_name":"Punggol","x":0.824989085928265,"y":0.208170302088758,"lon":103.9023059966562,"lat":1.405339576226845,"connections":[],"station_line":[6]},"93":{"id":93,"station_name":"Bukit Batok","x":0.164451762389254,"y":0.554378571086522,"lon":103.7497665649307,"lat":1.349520060732741,"connections":[{"next_station":110,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"94":{"id":94,"station_name":"Bukit Gombak","x":0.163601650003747,"y":0.49125772646262,"lon":103.752092995698,"lat":1.358695261269094,"connections":[],"station_line":[2]},"95":{"id":95,"station_name":"Yew Tee","x":0.164451762389254,"y":0.34110662637243,"lon":103.7473298800346,"lat":1.398067768060026,"connections":[{"next_station":96,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"96":{"id":96,"station_name":"Kranji","x":0.163601650003747,"y":0.267465640977878,"lon":103.7619406060964,"lat":1.424852644661057,"connections":[{"next_station":97,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"97":{"id":97,"station_name":"Marsiling","x":0.197606145424031,"y":0.183304514812676,"lon":103.7738204510363,"lat":1.432469522400065,"connections":[{"next_station":117,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"98":{"id":98,"station_name":"Admiralty","x":0.342975363345744,"y":0.165133362572462,"lon":103.8005424581603,"lat":1.440596285657938,"connections":[{"next_station":99,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"99":{"id":99,"station_name":"Sembawang","x":0.392281881705155,"y":0.165133362572462,"lon":103.8201097357584,"lat":1.448837319433129,"connections":[{"next_station":100,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"100":{"id":100,"station_name":"Canberra","x":0.444138737221088,"y":0.167046115439853,"lon":103.8292546439475,"lat":1.443511338844814,"connections":[{"next_station":101,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"101":{"id":101,"station_name":"Yishun","x":0.496845705122527,"y":0.185217267680067,"lon":103.8347836031844,"lat":1.429397167315948,"connections":[{"next_station":102,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"102":{"id":102,"station_name":"Khatib","x":0.524049301458754,"y":0.222515948594191,"lon":103.8332310200385,"lat":1.417869037116861,"connections":[{"next_station":103,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"103":{"id":103,"station_name":"Yio Chu Kang","x":0.533400537699332,"y":0.265552888110487,"lon":103.8446457604343,"lat":1.381953621306049,"connections":[{"next_station":104,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"104":{"id":104,"station_name":"Ang Mo Kio","x":0.532550425313825,"y":0.310502580494175,"lon":103.8498245724797,"lat":1.369882375268169,"connections":[{"next_station":111,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"105":{"id":105,"station_name":"Braddell","x":0.531700312928318,"y":0.393707330225681,"lon":103.847440830591,"lat":1.340672163943169,"connections":[{"next_station":106,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"106":{"id":106,"station_name":"Toa Payoh","x":0.531700312928318,"y":0.435787893308282,"lon":103.8477238043139,"lat":1.332376442030206,"connections":[{"next_station":107,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"107":{"id":107,"station_name":"Novena","x":0.509597390905134,"y":0.483606714993056,"lon":103.8433397334763,"lat":1.31989588603478,"connections":[{"next_station":112,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"108":{"id":108,"station_name":"Somerset","x":0.47559289548485,"y":0.629932309348464,"lon":103.8387655842814,"lat":1.300436220589932,"connections":[{"next_station":116,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"109":{"id":109,"station_name":"Marina South Pier","x":0.654966608826847,"y":0.927365380227758,"lon":103.8615038787278,"lat":1.273574804485931,"connections":[],"station_line":[2]},"110":{"id":110,"station_name":"Choa Chu Kang","x":0.16275153761824,"y":0.415703988200677,"lon":103.7445177987706,"lat":1.385736185403125,"connections":[{"next_station":95,"line_id":2,"connection_style":"solid"}],"station_line":[2]},"111":{"id":111,"station_name":"Bishan","x":0.530850200542811,"y":0.360234155046339,"lon":103.8502010819649,"lat":1.351053219933295,"connections":[{"next_station":11,"line_id":5,"connection_style":"solid"},{"next_station":105,"line_id":2,"connection_style":"solid"}],"station_line":[2,5]},"112":{"id":112,"station_name":"Newton","x":0.476443007870357,"y":0.518036266606093,"lon":103.8382423816355,"lat":1.313617269149807,"connections":[{"next_station":118,"line_id":2,"connection_style":"solid"},{"next_station":90,"line_id":4,"connection_style":"solid"}],"station_line":[2,4]},"113":{"id":113,"station_name":"Jurong East","x":0.163601650003747,"y":0.638539697251724,"lon":103.7426401629826,"lat":1.332526235648167,"connections":[{"next_station":93,"line_id":2,"connection_style":"solid"},{"next_station":66,"line_id":3,"connection_style":"solid"}],"station_line":[2,3]},"114":{"id":114,"station_name":"City Hall","x":0.604809978081928,"y":0.783908915173436,"lon":103.8518023338091,"lat":1.293454825951323,"connections":[{"next_station":115,"line_id":2,"connection_style":"solid"},{"next_station":115,"line_id":3,"connection_style":"solid"}],"station_line":[2,3]},"115":{"id":115,"station_name":"Raffles Place","x":0.605660090467435,"y":0.824076725388647,"lon":103.851621715866,"lat":1.285258281201247,"connections":[{"next_station":119,"line_id":2,"connection_style":"solid"},{"next_station":59,"line_id":3,"connection_style":"solid"}],"station_line":[2,3]},"116":{"id":116,"station_name":"Dhoby Ghaut","x":0.520648851916726,"y":0.680620260334325,"lon":103.8467879717855,"lat":1.298445092048172,"connections":[{"next_station":2,"line_id":5,"connection_style":"solid"},{"next_station":114,"line_id":2,"connection_style":"solid"},{"next_station":90,"line_id":6,"connection_style":"solid"}],"station_line":[2,5,6]},"117":{"id":117,"station_name":"Woodlands","x":0.286017833516769,"y":0.166089739006157,"lon":103.7865080837622,"lat":1.436673114687528,"connections":[{"next_station":122,"line_id":1,"connection_style":"solid"},{"next_station":98,"line_id":2,"connection_style":"solid"}],"station_line":[1,2]},"118":{"id":118,"station_name":"Orchard","x":0.444138737221088,"y":0.583069864097386,"lon":103.8306630989529,"lat":1.302857614748984,"connections":[{"next_station":131,"line_id":1,"connection_style":"solid"},{"next_station":108,"line_id":2,"connection_style":"solid"}],"station_line":[1,2]},"119":{"id":119,"station_name":"Marina Bay","x":0.604809978081928,"y":0.888153946446244,"lon":103.8544700329541,"lat":1.277353032977926,"connections":[{"next_station":25,"line_id":5,"connection_style":"solid"},{"next_station":135,"line_id":1,"connection_style":"solid"},{"next_station":109,"line_id":2,"connection_style":"solid"}],"station_line":[1,2]},"120":{"id":120,"station_name":"Sentosa","x":0.452639861076159,"y":0.995268107020137,"lon":103.820968627929,"lat":1.25555920600891,"connections":[],"station_line":[8]},"121":{"id":121,"station_name":"Woodlands North","x":0.248612888554457,"y":0.124009175923556,"lon":103.7854290806745,"lat":1.447936502496506,"connections":[{"next_station":117,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"122":{"id":122,"station_name":"Woodlands South","x":0.311521205081981,"y":0.194781032017022,"lon":103.7943652110204,"lat":1.425843386909275,"connections":[{"next_station":123,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"123":{"id":123,"station_name":"Springleaf","x":0.333624127105166,"y":0.222515948594191,"lon":103.8183906071381,"lat":1.397820804057506,"connections":[{"next_station":124,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"124":{"id":124,"station_name":"Lentor","x":0.357427273899364,"y":0.250250865171359,"lon":103.8358234520238,"lat":1.385890990086842,"connections":[{"next_station":125,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"125":{"id":125,"station_name":"Mayflower","x":0.38208053307907,"y":0.277985781748528,"lon":103.8367959780974,"lat":1.371276322901538,"connections":[{"next_station":126,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"126":{"id":126,"station_name":"Bright Hill","x":0.405033567487762,"y":0.305720698325697,"lon":103.8336354139939,"lat":1.364037182165615,"connections":[{"next_station":127,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"127":{"id":127,"station_name":"Upper Thomson","x":0.423736039968918,"y":0.354495896444167,"lon":103.8318330090042,"lat":1.355736574706553,"connections":[{"next_station":145,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"129":{"id":129,"station_name":"Napier","x":0.425436264739932,"y":0.516123513738703,"lon":103.8194837777762,"lat":1.306935987922338,"connections":[{"next_station":130,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"130":{"id":130,"station_name":"Orchard Boulevard","x":0.424586152354425,"y":0.55055306535174,"lon":103.8306630989529,"lat":1.302857614748984,"connections":[{"next_station":118,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"131":{"id":131,"station_name":"Great World","x":0.424586152354425,"y":0.624194050746292,"lon":103.833405504371,"lat":1.296552594564735,"connections":[{"next_station":132,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"132":{"id":132,"station_name":"Havelock","x":0.425436264739932,"y":0.664361860961502,"lon":103.8333695202785,"lat":1.288843563353537,"connections":[{"next_station":91,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"133":{"id":133,"station_name":"Maxwell","x":0.513847952832669,"y":0.807818326015823,"lon":103.8429163055429,"lat":1.280191522380937,"connections":[{"next_station":134,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"134":{"id":134,"station_name":"Shenton Way","x":0.56910525789063,"y":0.872851923507116,"lon":103.8518357073815,"lat":1.277462058335655,"connections":[{"next_station":119,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"135":{"id":135,"station_name":"Gardens by the Bay","x":0.78248346665291,"y":0.854680771266902,"lon":103.867154122142,"lat":1.278303439968121,"connections":[{"next_station":136,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"136":{"id":136,"station_name":"Tanjong Rhu","x":0.842841446023914,"y":0.787734420908218,"lon":103.8734779049167,"lat":1.296459488281608,"connections":[{"next_station":137,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"137":{"id":137,"station_name":"Katong Park","x":0.866644592818113,"y":0.762868633632136,"lon":103.885069217605,"lat":1.298379682791139,"connections":[{"next_station":138,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"138":{"id":138,"station_name":"Tanjong Katong","x":0.888747514841297,"y":0.735133717054967,"lon":103.897045454194,"lat":1.299038692777171,"connections":[{"next_station":139,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"139":{"id":139,"station_name":"Marine Parade","x":0.912550661635496,"y":0.708355176911494,"lon":103.9059103666293,"lat":1.302093293643494,"connections":[{"next_station":140,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"140":{"id":140,"station_name":"Marine Terrace","x":0.936353808429694,"y":0.680620260334325,"lon":103.9150201398566,"lat":1.306895321743126,"connections":[{"next_station":141,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"141":{"id":141,"station_name":"Siglap","x":0.961857179994907,"y":0.652885343757156,"lon":103.9287926248893,"lat":1.308850315558458,"connections":[{"next_station":142,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"142":{"id":142,"station_name":"Bayshore","x":0.98140976486157,"y":0.63088868578216,"lon":103.9412617256652,"lat":1.311898718691735,"connections":[{"next_station":143,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"143":{"id":143,"station_name":"Bedok South","x":1.00351268688475,"y":0.606979274939773,"lon":103.9294042796889,"lat":1.32407464586546,"connections":[{"next_station":144,"line_id":1,"connection_style":"solid"}],"station_line":[1]},"144":{"id":144,"station_name":"Sungei Bedok","x":1.03071628322098,"y":0.576375229061518,"lon":103.9294042796889,"lat":1.32407464586546,"connections":[],"station_line":[1,4]},"145":{"id":145,"station_name":"Caldecott","x":0.425436264739932,"y":0.399445588827854,"lon":103.8403874992376,"lat":1.338259729941808,"connections":[{"next_station":22,"line_id":5,"connection_style":"solid"},{"next_station":50,"line_id":1,"connection_style":"solid"}],"station_line":[1,5]},"146":{"id":146,"station_name":"Xilin","x":1.05555,"y":0.54,"lon":103.9649,"lat":1.3289,"connections":[{"next_station":144,"line_id":4,"connection_style":"solid"}],"station_line":[4]},"147":{"id":147,"station_name":"Hume","x":0.249463000939964,"y":0.383333,"lon":103.7692,"lat":1.3544,"connections":[{"next_station":29,"line_id":4,"connection_style":"solid"}],"station_line":[4]}}}'

run()
```
